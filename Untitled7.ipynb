```python
import numpy as np
```

```python
def gradiant_configration(x,y):
  m = 0
  b = 0
  n =len(x)
  iter = 1000
  lr = 0.08
  for i in range(iter):
    y_pred = m*x+b
    md = -(2/n)*sum(x*(y-y_pred))
    bd = -(2/n)*sum(y-y_pred)
    m = m -lr*md
    b = b -lr*bd
    print(f"m = {m} b = {b} i = {i}")
  return(m,b)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradiant_configration(x,y)
```

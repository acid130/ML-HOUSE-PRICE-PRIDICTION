```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/content/house prices.csv')
```

```python
df
```

```python

plt.scatter(df.area,df.price,color='red',marker='+')
plt.xlabel("area")
plt.ylabel("price")
plt.show
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[2722.51]])
```

```python
plt.scatter(df.area , df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.xlabel("area")
plt.ylabel("price")
plt.show
```
